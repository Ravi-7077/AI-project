```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
crew=pd.read_csv('tmdb_5000_credits.csv')
```

```python
crew.head()
movies.head()
```

```python
movies=movies.merge(crew,on='title')
```

```python
movies.head()
```

```python
movies=movies[['id','title','crew','overview','cast','genres','keywords']]
```

movies.head()

```python
movies.head()
```

```python
movies.dropna()
```

```python
import ast
def fn(oj):
    li=[]
    for i in ast.literal_eval(oj):
        li.append(i['name'])
    return li
```

```python
movies['genres']=movies['genres'].apply(fn)
```

```python
movies['keywords']=movies['keywords'].apply(fn)
```

```python
import ast
def fn3(oj):
    li=[]
    count=0
    for i in ast.literal_eval(oj):
        if count!=5:
            li.append(i['name'])
            count+=1
        else:
            break
        
    return li
```

```python
movies['cast']=movies['cast'].apply(fn3)
```

```python
movies.head()
```

```python
movies['cast'].iloc[0]
```

```python
import ast
def fn5(oj):
    li=[]
    for i in ast.literal_eval(oj):
        if i['job']=='Director' or i['job']=='Producer':
            li.append(i['name'])
    return li
```

```python
movies['crew']=movies['crew'].apply(fn5)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['tags'] = movies['cast'] + movies['crew'] + movies['genres'] + movies['keywords']
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:str(x).split())
```

```python

movies['tags'] = movies['cast'] + movies['crew'] + movies['genres'] + movies['keywords']+movies['overview']
```

```python
reqdata=movies[['id','title','tags']]
```

```python
reqdata['tags']=reqdata['tags'].apply(lambda x:" ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=vec.fit_transform(reqdata['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return  " ".join(y)
        
```

```python
reqdata['tags'].apply(stem)
```

```python
reqdata['tags']=reqdata['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similardist=cosine_similarity(vector)
```

```python
list(enumerate(similardist))
```

```python
sorted(list(enumerate(similardist[0])),reverse=True,key=lambda x:x[1])[1:7] #sort on the basis of dist
```

```python
def recommend(movie):
    movie_index=reqdata[reqdata['title']==movie].index[0]
    distance=similardist[movie_index]
    movielist=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:7]
    
    for i in movielist:
        print(i[0])
        print(reqdata.iloc[i[0]].title)
       
```

```python
recommend("Batman Begins")
```
